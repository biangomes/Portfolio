# 1. Introducao

A probabilidade pode ser vista como uma forma de quantificar a incerteza associada a eventos a partir do universo deles. Suponha que você queira jogar um dado. O universo são todos os valores possíveis, porém ao cair e dar um número, um evento associado ao universo está acontecendo. Suponha que caiu o número 4, portanto, é um evento; número 5, outro evento; e assim por diante.

A notação matemática que indica a probabilidade de um evento $E$ ocorrer é:
$$ P(E) $$



# 2. Dependência e Independência

Dado dois eventos, $E$ e $F$, diz-se que eles são **dependentes** se um deles ocorre e isso implica na ocorrência do outro. Caso contrário, são **independentes**.

Imaginando o cenário do lançamento de uma moeda:

> Suponha que o evento $E$ seja previamente conhecido: a primeira jogada é **coroa**. <br>
> Ao lançar novamente a moeda, não é possível prever se o evento $F$ será **cara** ou **coroa** novamente.
> Portanto, são **independentes**.

Contudo, se quisermos saber se **ambas** as jogadas foram **cara**, naturalmente não foram. Pois, como dito anteriormente, o evento $E$ (previamente conhecido) resultou em **coroa**. Esses dois eventos, agora, são **dependentes**.

Matematicamente, dizemos que dois eventos, $E$ e $F$, são independentes se a probabilidade deles acontecerem é o produto da probabilidade de que cada um deles aconteça:

$$ P(E, F) = P(E).P(F) $$

# 3. Probabilidade Condicional

Quando os dois eventos $E$ e $F$ são independentes, por definição temos: $ P(E, F) = P(E).P(F) $. 

Se eles não são necessariamente independentes e $P(F) \neq 0$, então definimos a probabilidade de $E$ "condicionada a $F$" assim:

$$ P(E|F) = \frac {P(E,F)}{P(F)} $$
<br>
$$ \rightarrow P(E,F) = P(E|F).P(F) $$



# 4. Teorema de Bayes

O Teorema de Bayes é uma maneira de "reverter" a probabiliade condicional. Como se você precisasse saber a probabilidade de algum evento $E$ ser condicionado à ocorrência de um evento $F$, com a **única** informação de que a **probabilidade da ocorrência de $F$ depende da ocorrência e $E$**.


# 5. Distribuição Normal

A distribuição normal é a distribuição mais conhecida. A sua forma é uma curva em forma de sino e possui dois parâmetros:

- $\mu$ (média): indica **onde** o sino está **centralizado**;

- $\sigma$ (desvio padrão): indica a **largura** do sino.

Matematicamente:

$$ f(x|\mu,\sigma) = \frac {1}{\sqrt{2.\pi.\sigma}}.exp- \frac {(x-\mu)^2} {2\sigma^2} $$

A célula abaixo faz a implementação da distribuição normal.

```
# Criando a função da distribuicao normal

import math as m

def normal_pdf(x, mu=0, sigma=1):
  sqrt_dois_pi = m.sqrt(2 * m.pi)
  return (m.exp(-(x-mu)**2 / 2 / sigma**2) / (sqrt_dois_pi * sigma))
```

```
# Plotando varias curvas da funcao normal_pdf
import matplotlib.pyplot as plt

xs = [x/10.0 for x in range(-50, 50)]
plt.plot(xs, [normal_pdf(x, sigma=1) for x in xs], '-', label='mu=0, sigma=1')
plt.plot(xs, [normal_pdf(x, sigma=2) for x in xs], '--', label='mu=0, sigma=2')
plt.plot(xs, [normal_pdf(x, sigma=0.5) for x in xs], ':', label='mu=0, sigma=0.5')
plt.plot(xs, [normal_pdf(x, mu=-1) for x in xs], '-.', label='mu=-1, sigma=1')
plt.legend()
plt.title("Diversas funções de Densidade de Probabilidade Normais")
plt.show();
```
