# 1. Introdução
De acordo com [Portal Action](http://www.portalaction.com.br/series-temporais/introducao), uma série temporal é um conjunto de observações feitas sequencialmente ao longo do tempo. O comportamento dessas séries são descritos por processos estocásticos, ou seja, processos que são controlados por leis probabilísticas. 

Algumas características dessa série são:

- Observações vizinhas são **dependentes**;

- Os dados são **ordenados**;

- A distribuição do tempo pode não ser explícita, isto é, pode estar na forma de espaço, profundidade, dentre outros.

Através das séries temporais, podemos **identificar padrões** de acontecimentos e/ou **fazer previsões**.

### 1.1 Componentes de uma série temporal

É comum encontrar componentes de séries temporais, como:

- **tendência**;

- **sazonalidade**;

- **comportamento ciclíco**.


#### 1.1.1 Tendência

A tendência é identificada em padrões de crescimento ou decrescimento, podendo ser linear ou não. Pode-se ter, também, mudanças de direção ao longo do tempo. Isto é, por um período o gráfico estava crescente e, em algum momento, passou-se a identificar uma tendência decrescente.


#### 1.1.2 Sazonalidade

A sazonalidade é uma componente que tem os períodos fixos e são de conhecimento do analista. Existe sazonalidade qunado a série é influenciada por fatores que acontecem diaria, semanal, mensal e/ou trimestralmente.


#### 1.1.3 Comportamento cíclico

Diferente da sazonalidade, o comportamento cíclico não tem período fixo e não se tem conhecimentos prévios sobre eles. A duração das flutuações são **maiores** que as sazonais.

### 1.2 Natureza Matemática da Série Temporal

Uma série temporal possui até duas naturezas: 

- **Contínua** no tempo, sem "buracos". 
Considere um conjunto $ T = {t : t_{1} < t < t_{2}} $. A série temporal é uma função de $T $, $ { X(t) : t \in T } $.

- **Discreta** no tempo, em que são em **pontos específicos**. Considere um conjunto $ T \in \mathbb{N^{*}} | T = {1, 2, ..., n} $. A série temporal é dada por: $ X_{t} : t \in T $.

O que foi dito até então compreende a série temporal **univariada**, em que apenas uma **única** sequência de dados é analisada. Quando se tem **diversas coleções** de dados para a mesma sequência, trata-se, então, de uma análise **multivariada**.

Como dito no tópico 1, as séries temporais são o resultado de algumas componentes: componente sazonal ($S_{t}$), tendência-cíclica ($T_{t}$) e um resíduo ($E_{t}$, o que "resta").

Podemos decompor uma série temporal em dois modelos: aditivo e multiplicado.

#### 1.2.1 Decomposição aditiva

A decomposição aditiva é dada da seguinte forma:

$$ y_{t} = S_{t} + T_{t} + E_{t} $$

Esta decomposição é mais apropriada quando as componentes **não variam com a magnitude dos dados**.


#### 1.2.2 Decomposição multiplicativa

A decomposição multiplicativa é dada da seguinte forma:

$$ y_{t} = S_{t} x T_{t} x E_{t} $$

Esta decomposição é mais apropriada quando a variação das componentes é **proporcional à magnitude**.
